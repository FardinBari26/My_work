```python

import numpy as np
# Exercise 1 — NumPy

"""Given `arr = np.random.randint(1, 101, 20)`:
1. Print its sum, min, and max.
2. Create a second array of the same size and multiply them element-wise."""

arr = np.random.randint(1, 101, 20)
print("First array :", arr)
print("Sum :", arr.sum())
print("Min :", arr.min())
print("Max :", arr.max())

# creating sewcoond arr

arr2 = np.random.randint(1, 101, 20)

# multiply them element-wise

new_array = arr * arr2
print("Second array :", arr2)
print("Element-wise multiplication :", new_array)
```

```python
# Exercise 2 — Pandas
"""Create a 5×3 DataFrame of random integers (1–100) with columns A, B, C. Print the
sum and mean of each column."""
# NumPy ↔ Pandas: they share memory layout, so conversion is trivial
import pandas as pd
import numpy as np

# Create a 5×3 DataFrame of random integers from 1 to 100
df = pd.DataFrame(
    np.random.randint(1, 101, size=(5, 3)),
    columns=["A", "B", "C"]
)

print("DataFrame:")
print(df)

# Print sum of each column
print("\nColumn sums:")
print(df.sum())

# Print mean of each column
print("\nColumn means:")
print(df.mean())
```
